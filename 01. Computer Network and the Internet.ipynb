# Chapter 01. Computer Network and Internet
## 컴퓨터 네트워크와 인터넷
------------
Goal:
1. 컴퓨터 네트워킹과 인터넷의 개요를 안다.
2. 컴퓨터 네트워크의 큰 그림을 그림으로 숲에서 나무로 가는 방향을 잡는다.

### 1.1 인터넷이란 무엇인가?
#### 1.1.1 구성요소로 본 인터넷
> 전 세계적으로 수십억 개의 컴퓨팅 장치를 연결하는 컴퓨터 네트워크

* Host / end System : 인터넷에 연결되는 컴퓨팅 장치(PC, laptop, TV, ...)
* 종단 시스템은 통신 링크(Communication link)와 패킷 스위치(Packet S/W)의 네트워크로 연결
    - 통신 링크
        + 동축 케이블, 구리선, 광케이블, 라디오 스펙트럼 등의 물리매체(Physical media)로 구성
        + 다양한 전송률(transmission rate)를 이용
        + 단위: [bps]
    
    - 패킷 스위치
        + 교차로, 고속도로 등의 수송네트워크와 유사
        + 라우터(Router), 링크계층스위치(link-layer s/w)가 대표적
* Protocol: 라우터와 종단 시스템 간 송수신되는 패킷 포맷 기술
    - ex) TCP/IP
* 인터넷 표준
    - IETF -> RFCs: TCP, IP, HTTP, SMTP, ...
    - IEEE -> 802 LAN/MAN: Ethernet, Wireless WIFI, ...

#### 1.1.2 서비스 측면에서 본 인터넷
> Application에 서비스를 제공하는 인프라 구종

* Application: 많은 종단 시스템을 포함하기에 분산 애플리케이션(distributed application)이라 부름
* Socket Interface: 송신 프로그램이 따라야 하는 규칙의 집합
    - ex) Alice ----mail----> Bob
            letter -> envelope -> Name&Addr. -> Stamp -> Post office

#### 1.1.3 프로토콜이란 무엇인가?
> Protocol: 둘 이상의 통신 개체 간의 교환되는 메시지 포맷, 순서, 송수신 등의 이벤트에 따른 행동을 정의한 것

![1.1.3](https://t1.daumcdn.net/cfile/tistory/2661BA4958E0448725)

* in Human
바른 예절로 시간을 묻는 경우,
    1. 인사를 통해 커뮤니케이션을 요청
    2. 커뮤니케이션을 수락 했다는 응답을 반환
    3. 시간을 묻는다는 목적을 수행
    4. 상대방의 목적 행동에 맞는 피드백을 반환

* in Network
어떠한 연결을 원하는 요청하고, 그에 따른 응답을, 목적행동을 수행하고, 그에 대한 맞는 피드백을 반환

### 1.3 네트워크 코어
> What is Mesh

#### 1.3.1 패킷 스위칭(Packet Switching)
송신측(src) -> 메시지(msg, img, ...) -> 패킷(packet) 단위로 분할 -> __패킷스위치__ -> 수신측(dst)

* 패킷 스위치
    + 저장-후-전달 전송(Store-and-Forward transmission) 방식
    
    패킷 스위치(라우터 등)가 전체 패킷을 받아서 저장후 이를 다시 송신하는 방식
    ![saft img](http://bucarotechelp.com/networking/basics/images/figure_1_11.jpg)
    ex) each Packet size : L, R bps, 3 packets
    |time|src|-->|Router|-->|dst|
    |:--:|:--:|:--:|:--:|:--:|:--:|
    |0|[1,2,3]|[1]|
    |L/R|[2,3]|[2]|[1]|[1]
    |2L/R|[3]|[3]|[2]|[2]|[1]
    |3L/R|||[3]|[3]|[1,2]
    |4L/R|||||[1,2,3]
    
    if) src ---N link---> dst, R bps, Packet size = L
    
    $d_종단간지연 = N \times$ $L\over R$
    
    
    + 큐잉 지연과 패킷 손실
    
        1. 패킷 스위치는 여러 링크를 보유
        2. 각 링크에 대해 출력 버퍼(출력 큐)를 보유
        3. 링크가 다른 패킷을 전송 중이라면, 버퍼에 도착하는 패킷은 출력 버퍼에서 대기
        -> 큐잉 지연(queuing delay) 발생 : 가변적(네트워크 혼잡도 등에 영향을 받음)
        4. 버퍼가 전송 대기 중인 패킷들로 가득 차 있을 경우, 도착하는 패킷은 공간이 없어 손실되거나, 큐에 대기 중인 패킷을 폐기(drop)
        -> 패킷 손실(packet loss) 발생
        
    + 전달 테이블과 라우팅 프로토콜
    
> 라우터는 어떻게 특정 패킷을 어느 링크로 전달하는지 판단하는가?
    
        1. Packe의 header에는 dst IP addr.가 포함됨
        2. 라우터는 dst addr.를 라우터의 출력 링크로 맵핑하는 __전달 테이블(forwarding table)__ 을 갖고 있음
        3. 패킷이 라우터에 도착하면, 라우터는 올바른 출력 주소를 조사하고 목적지 주소를 이용하여 전달 테이블을 검색, 해당 링크로 전송한다.
        
> 전달 테이블은 어떻게 설정되는가? => 라우팅 프로토콜
    
#### 1.3.2 회선 교환(Circuit Switching)
링크와 스위치의 네트워크를 통한 데이터를 이동시키는 방식
1. 회선교환(circuit switching)
2. 패킷교환(packet switching) -> 1.3.1

* 개념
    + 종단 간 통신을 위해 경로 상 필요한 자원(버퍼, 링크전송률)을 세션(session) 동안 예약(reserve)
    + 통신 링크 접속을 위한 대기가 있을 수 있음
    + 회선(circuit): 송수신자 간 경로에 있는 스위치들이 해당 연결 상태를 유지해야 하는 연결
    + 연결 중에는 네트워크 링크에 일정 전송률을 보장 -> 보장된(guaranteed) 일정 전송률로 통신 가능
    + 종단간 연결(end-to-end connection)

* 다중화
    1. 주파수-분할 다중화(FDM, Frequency-division multiplexing)
        - 링크의 주파수 스펙트럼 공유
        - 일반적으로 4kHz의 주파수대역폭(__bandwidth__) 을 가짐
    
    2. 시-분할 다중화(TDM, Time-division multiplexing)
        - 시간을 일정 주기의 '프레임(frame)'으로 구분
        - 고정된 수의 시간 '슬롯(slot)'으로 나눔
    
* 패킷교환 vs 회선교환
    + 회선교환 Cons
        - 할당된 회선이 노는 시간, 비활용기간(silent period)로 인해 비효율적임
    
    + 패킷교환 Cons
        - 가변적, 예측 불가의 종단간 지연(대부분 큐잉지연) => 실시간 서비스에 부적합하다는 주장
    
    + 효율성 측면
        - 패킷교환 방식이 회선교환 방식보다 전송 용량 고유에 효율적
        - 패킨교환 방식은 간단하며, 구현 비용이 적음
        
    + 패킷 교환의 효율성 설명(간단한 예시)
    
        - if) 10명의 사용자, 한 사용자가 한번에 1000bit 패킷을 1000개 생성하고 다른 사용자가 패킷을 생성하지 않을 때,
            1. circuit switching - TDM(1 frame = 10 slot, 1 slot = 1000 bit)
            
            한 프레임 당 한 개의 시간 슬롯만 사용할 수 있는데, 나머지 9개 슬롯은 쉬는 상태가 됨. 이 경우 데이터 100만 비트를 모두 전송하기 위해서는 10초가 걸림.
            2. packet switching
            
            패킷을 생성하는 다른 사용자가 없기에 다중화가 요구되지 않고, 1Mbps의 링크가 가득 찰 때까지 패킷을 계속 보낼 수 있음. 즉, 사용자의 데이터는 1초만에 전송됨.
